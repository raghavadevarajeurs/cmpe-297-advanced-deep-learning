## Model Hyperparameters

```python
target = 'Survived'

keras_model_type = "auto"
project_name = "titanic"
model_options = {'nlp_char_limit':50, 'cat_feat_cross_flag':"",
                 'max_trials': 5, "tuner": "storm"}
keras_options = {"patience":10, 'class_weight': True, 'early_stopping': True, 
                 'lr_scheduler': '', "optimizer": 'RMS'}
```

```python
model, cat_vocab_dict = deepauto.fit(train, target, keras_model_type,project_name=project_name, keras_options=keras_options,
                                     model_options=model_options, save_model_flag=False, use_my_model='', verbose=1)
```

## Model Prediction

```python
predictions = deepauto.predict(model, project_name, test_dataset=test, 
                               keras_model_type=keras_model_type, cat_vocab_dict=cat_vocab_dict)
```
